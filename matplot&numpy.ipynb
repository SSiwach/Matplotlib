```python
import numpy as np

n = 100

x = np.random.normal(size = n)

print(np.mean(x))


y = 4 + 3*x + np.random.normal(size = n)

print(np.mean(y))

z = x**3 + np.random.normal(size=n)

print(np.mean(z))

W = np.vstack((x,y,z))

Sigma = np.cov(W)

print(Sigma)

Eigvals, Eigvecs = np.linalg.eig(Sigma)

print(Eigvals)
```

```python
import numpy as np

import pandas as pd

from IPython.display import display

np.random.seed(100)

rows = 10
cols = 5

a = np.random.randn(rows, cols)

df = pd.DataFrame(a)

df = pd.DataFrame(a)

display(df)


print(df.mean())

print(df.std())

display(df**2)
```
