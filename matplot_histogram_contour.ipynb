```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = [1,3,5,2,4,4,2,5,4,5,4,3,1,2]

n_bins = 5

plt.hist(x, bins = n_bins)
plt.show()
```

```python
import numpy as np

np.random.seed(31415)

n_points = 10000

n_bins = 15

x = np.random.randn(n_points)

plt.hist(x, bins = n_bins )

plt.show
```

```python
y = np.random.randn(n_points)

plt.hist2d(x, y, bins = 50)

plt.show()
```

```python
plt.hist(x,20,density = True, histtype = 'stepfilled', facecolor = 'g', alpha = 0.5)

plt.show()
```

```python
plt.hist(x, 20, density = True, histtype = 'step')

plt.show()
```

```python
x = np.arange(-3, 3, 0.005)

y = np.arange(-3, 3, 0.005)

x, y = np.meshgrid(x,y)

z = (x**2 + y**2)

out = plt.contour(x, y, z)

plt.clabel(out, inline = True, fontsize = 10)

plt.show()
```

```python
out = plt.contour(x,y,z)

plt.clabel(out, inline = True, fontsize=
           10)

plt.colorbar(out)

plt.show()
```

```python
out = plt.contour(x, y, z, colors = 'g')

plt.clabel(out, inline = True, fontsize = 10)

plt.show()
```

```python
plt.contourf(x, y, z, hatches = ['-','/','\\','//'],cmap = 'cool', alpha = 0.75)

plt.show()
```

```python
y, x = np.mgrid[-5:5:200j, -5:5:300j]

u = x**2 + y**2

v = x + y

plt.streamplot(x, y, u, v)

plt.show()
```

```python
plt.streamplot(x, y, u, v, density = [0.5, 0.75])
```

```python
plt.streamplot(x,y,u,v, color = u, linewidth = 1, cmap = 'cool')

plt.show()
```

```python
plt.streamplot(x, y, u, v, density = 0.6, color = 'k', linewidth = x)

plt.show()
```

```python
import numpy as np

X = np.arange(-5, 5, 0.5)
Y = np.arange(-10, 10, 1)

U, V = np.meshgrid(X, Y)

plt.quiver(X,Y,U,V)

plt.show()
```

```python
img3 = [[1,2,3],[3,4,5],[4,5,6]]

plt.imshow(img3)

plt.show()
```

```python
methods = ['none','antialiased','nearest','bilinear','bicubic','spline16','hanning','hamming','hermite','kaiser','quadric','catrom','gaussian','bessel','mitchell','sinc','lanczos','blackman']

fig, axs = plt.subplots(nrows = 4, ncols =6, figsize = (9,6), subplot_kw = {'xticks':[], 'yticks':[]})

for ax , interp_method in zip(axs.flat, methods):
  ax.imshow(img3, interpolation = interp_method, cmap = 'hot')

  ax.set_title(str(interp_method))

plt.tight_layout()

plt.show()
```
